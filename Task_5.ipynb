```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('deliveries.csv')#Reading the dataset
```

```python
data.shape
```

```python
sns.heatmap(data.isnull(),cmap = "coolwarm")
plt.show()
```

```python
data.columns
```

```python
data['batting_team'].value_counts()
```

```python
sns.barplot(x='batting_team',y='wide_runs',data=data)
plt.xticks(rotation=90)
plt.show()
```

__Here from the above barplot we can see that Kochi Tuskers Kerala has the Highest run rates__

```python
sns.barplot(x='batting_team',y='extra_runs',data=data)
plt.xticks(rotation=90)
plt.show()
```

__Here from the above barplot we can see that Kochi Tuskers Kerala has the Highest extra run rates__

```python
MI = data[data['batting_team']=="Mumbai Indians"]
```

```python
np.average(MI[MI["batsman"]=="SR Tendulkar"]["total_runs"])
```

```python
plt.figure(figsize=(16,4))
sns.barplot(x='batsman',y='total_runs',data=MI)
plt.xticks(rotation=90)
plt.show()
```

__The above barplot indicates the run rate per ball of the Batsman in the 'Mumbai Indians' team. And the black line at the top of every bar shows the variance of the run rate of individual player in the team.

```python
plt.figure(figsize=(16,4))
sns.barplot(x='over',y='total_runs',data=MI,color='lightblue')
plt.xticks(rotation=90)
plt.show()
```

__Here we can analyse that 'mumbai indians' performs well in the last overs comapred to the first and middle overs__

```python
plt.figure(figsize=(16,4))
sns.barplot(x='batting_team',y='total_runs',data=data, palette='Blues')
plt.xticks(rotation=90)
plt.show()
```

__The run rate of the 'Delhli Capitals' team is strong and the teams like 'Chennai Super Kings' and 'Rising Pune Supergiants' are really competitive__

```python
plt.figure(figsize=(16,4))
sns.barplot(x='over',y='total_runs',data=data[data['batsman']=='MS Dhoni'],color='pink')
plt.xticks(rotation=90)
plt.show()
```

__If we see the Stratergy of MS. Dhoni we can really bet on this person, because this person is the very stable player in all the overs and the run rate keeps on increasing all the way till the end.
